1) Resolva as equações.
Itens (b), (c) e (f) 

```python
'''
b) |5x − 3| = |3x + 5|

Para equações modulares, defini primeiro, as possibilidades dos sinais
da equação, com as variaveis eq_x, posteriormente, fiz um solve em ca-
da uma das equações, guardando nas variáveis resul_x, depois,verifiquei 
se existiam soluções iguais, para exibir unicamente cada resposta. Por
fim exibi-as com o 'show'.
Idem para todos os outros itens do Exércicio 1.
'''

eq_a = 5*x - 3 == 3*x + 5
eq_b = 5*x - 3 == -(3*x + 5)
eq_c = -(5*x - 3) == 3*x + 5
eq_d = -(5*x - 3) == -(3*x + 5)

resul_a = solve(eq_a, x)
resul_b = solve(eq_b, x)
resul_c = solve(eq_c, x)
resul_d = solve(eq_d, x)

all_resul = resul_a + resul_b + resul_c + resul_d
resul = []
for res in all_resul : 
    if res not in resul : 
        resul.append(res)

show(resul)
```

```python
'''
c) |7x| = 4 − x

Neste caso, não foi necessário a verificação de respostas iguais, pois
não têm módulos dos dois lados da equação.
'''
eq_a = 4 - x == 7*x
eq_b = 4 - x == -(7*x)

resul_a = solve(eq_a, x)
resul_b = solve(eq_b, x)

all_resul = resul_a + resul_b
show(all_resul)
```

```python
'''
f) |x + 2| / |x − 2| = 5

Para este item, adotei primeiro a estratégia de 'multiplicar em cruz'
ou seja, multipliclar o denominador da fração por 5, facilitando o 
entendimento da equação.
Equação: |x + 2| = 5 . |x - 2|
'''

eq_a = x + 2 == 5 * (x - 2)
eq_b = -(x + 2) == 5 * (x - 2)
eq_c = x + 2 == -5 * (x - 2)
eq_d = -(x + 2) == -5 * (x - 2)

resul_a = solve(eq_a, x)
resul_b = solve(eq_b, x)
resul_c = solve(eq_c, x)
resul_d = solve(eq_d, x)

all_resul = resul_a + resul_b + resul_c + resul_d
resul = []

for res in all_resul:
    if res not in resul:
        resul.append(res)

show(resul)
```
